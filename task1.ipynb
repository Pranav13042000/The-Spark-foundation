**Problem Statement** :Predict the percentage of an student based on the no. of study hourse.What will be predicted score if a student studies for 9.25 hours/day?






```python
# importing important libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Import dataset
url = "http://bit.ly/w-data"
df  = pd.read_csv(url)
```

```python
df.head()
```

```python
# create basic scatterplot target variable and feature
plt.plot(df['Hours'],df['Scores'],'o')
plt.title('Hourse vs Scores')
plt.xlabel('Hours Studied')
plt.ylabel('percentage Score')
plt.show()
```

```python
df.corr()
```

```python
sns.heatmap(df.corr(),cmap="YlGnBu",linewidths=0.1)
```

**As we can see from above abservation there data**

```python
# splitting target variable and feature
x =df.iloc[:,:-1].values
y =df.iloc[:,:1].values
```

```python
x.shape
```

```python
y.shape
```

```python
# split data into train and 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
# Training liner model
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
# predicting train and test Values
test_pred=lr.predict(x_test)
train_pred=lr.predict(x_train)
```

```python
# plotting for the train data 
plt.scatter(x_train,y_train)
plt.plot(x_train,train_pred,color='red')
plt.show()
```

```python
# plotting for the test data
plt.scatter(x_test,y_test)
plt.plot(x_test,test_pred,color='red')
plt.show()
```

```python
from sklearn import metrics
print('Mean Absolute Error:',
      metrics.mean_absolute_error(y_test ,test_pred))
      
```

```python
# model Score on train data
r2_score =lr.score(x_train,y_train)
print(r2_score*100,'%')
```
